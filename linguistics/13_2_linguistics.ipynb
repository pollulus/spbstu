```python
import pandas as pd
import yake
```

```python
def find_keywords(file_name):
    df = pd.read_csv(file_name, encoding='utf-8')

    kw_extractor = yake.KeywordExtractor(lan='ru', n=2, top=5)

    df['key_words'] = df['normalized_text'].apply(kw_extractor.extract_keywords)    df.to_csv('cafe_reviews_with_keywords.csv', index=False, encoding='utf-8')
    return df

find_keywords('cafe_reviews.csv')
```
