```python
import pandas as pd
import matplotlib.pyplot as plt

#  DataFrame'   ( ,   2023  2024 )
df = pd.read_csv('books.csv', encoding='utf-8')
df = df.copy()

#    
df_2023 = df[df[''] == 2023]
df_2024 = df[df[''] == 2024]

#   (-     2023 .)
def chart_2023(df_2023):
    plt.plot(df_2023[''], df_2023[''],
             #  -,    ,  
             marker='o', color='red', label=' ')
    plt.title('  2023') # 
    plt.xlabel('') #   x
    plt.ylabel(' ') #   y
    plt.grid(True) # 
    plt.legend()  #  


#   (-     2024 .)
def chart_2024(df_2024):
    plt.bar(df_2024[''], df_2024[''],
            #    ,  
            color='green', label=' ')
    plt.title('  2024')  # 
    plt.xlabel('')  #   x
    plt.ylabel(' ')  #   y
    plt.grid(True)  # 
    plt.legend()  #  

#   
plt.subplot(1, 2, 1)  #     
chart_2023(df_2023)  #  

#   
plt.subplot(1, 2, 2)
chart_2024(df_2024)

plt.show()
```
