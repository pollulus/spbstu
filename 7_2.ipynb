```python
def check_input(item):
    if type(item) == str:
        return True
    else:
        return False


def get_substring(input_string):
    longest_substring = ""
    building_substring = ""

    for i in input_string:
        if i in building_substring:
            # Поиск позиции первого вхождения символа
            position = building_substring.index(i)
            # Срез: все после первого вхождения символа
            building_substring = building_substring[position + 1:] + i
        else:
            building_substring = building_substring + i

        if len(building_substring) > len(longest_substring):
            longest_substring = building_substring

    return longest_substring


def main(input_list):
    longest_string_index = -1
    longest_string = ""
    current_index = 0

    for item in input_list:
        if check_input(item):
            if len(item) > len(longest_string):
                longest_string = item
                longest_string_index = current_index

        current_index += 1

    if longest_string_index == -1:
        print("-1, ''")
        return

    longest_substring = get_substring(longest_string)
    print(f"{longest_string_index}, '{longest_substring}'")
```
